```python
train_pos=train_raw[train_raw['target']==1]
print train_pos['target'].value_counts()
train_neg=train_raw[train_raw['target']==0]
print train_neg['target'].value_counts()
```

```python
#拆分数据集
from sklearn.cross_validation import train_test_split  

#df转narray
train_pos=np.array(train_pos)
train_neg=np.array(train_neg)
#数据集拆分
train_pos_new,test_pos_new = train_test_split(train_pos,  
                                                   test_size = 0.2,  
                                                   random_state = 0)  
train_neg_new,test_neg_new = train_test_split(train_neg,  
                                                   test_size = 0.2,  
                                                   random_state = 0)  
#数据集拼接
train_new=np.concatenate([train_pos_new,train_neg_new],axis=0) #在纵轴上合并
print train_new.shape

test_new=np.concatenate([test_pos_new,test_neg_new],axis=0) #在纵轴上合并
print test_new.shape
  
```

```python
train_temp=pd.DataFrame(train_new)
train_temp.columns=train_raw.columns
train=train_temp

test_temp=pd.DataFrame(test_new)
test_temp.columns=train_raw.columns
test=test_temp
```

## Define a function for modeling and cross-validation

This function will do the following:
1. fit the model
2. determine training accuracy
3. determine training AUC
4. determine testing AUC
5. update n_estimators with cv function of xgboost package
6. plot Feature Importance 

# 不调整样本比例

```python
predictors = [x for x in train.columns if x not in [target, IDcol]]
X = train[predictors]
y = train['target']
X_test = test[predictors]
y_test = test['target']
```

```python
from sklearn.metrics import roc_auc_score
def gini(y_actual, y_pred):
  return 2*roc_auc_score(y_actual, y_pred)-1
```

```python
gbm0 = GradientBoostingClassifier(random_state=10)
gbm0.fit(X,y)
```

```python

def model(gbm0,X,y,X_test,y_test):
    #训练集评测
    y_pred = gbm0.predict(X)
    y_predprob = gbm0.predict_proba(X)[:,1]
    print "Accuracy : %.4g" % metrics.accuracy_score(y.values, y_pred)
    print "AUC Score (Train): %f" % metrics.roc_auc_score(y, y_predprob)
    #GINI score
    print "GINI Score (Train): %f" % gini(y, y_predprob)

    #测试集评测
    y_test_pred = gbm0.predict(X_test)
    y_test_predprob = gbm0.predict_proba(X_test)[:,1]

    print "Accuracy : %.4g" % metrics.accuracy_score(y_test.values, y_test_pred)
    print "AUC Score (Test): %f" % metrics.roc_auc_score(y_test, y_test_predprob)
    #GINI score
    print "GINI Score (Test): %f" % gini(y_test, y_test_predprob)
    
model(gbm0,X,y,X_test,y_test)
```

# GridSearch 调参

```python
X2 = train[predictors]
y2 = train['target']
X2_test = test[predictors]
y2_test = test['target']
```

```python
#Grid seach on subsample and max_features
#Choose all predictors except target & IDcols

param_test1 = {'n_estimators':range(20,201,10)}
gsearch1 = GridSearchCV(estimator = GradientBoostingClassifier(learning_rate=0.1, subsample=0.8,random_state=10), 
                       param_grid = param_test1, scoring='roc_auc',iid=False,cv=5, n_jobs=-1)
gsearch1.fit(X2,y2)
gsearch1.grid_scores_, gsearch1.best_params_, gsearch1.best_score_
```

```python
estimator = GradientBoostingClassifier(n_estimators=200,learning_rate=0.1, random_state=10)
estimator.fit(X2,y2)
```

```python
model(estimator,X2,y2,X2_test,y2_test)
```
