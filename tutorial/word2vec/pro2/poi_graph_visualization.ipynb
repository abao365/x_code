```python
alist=[2004491,2004446,2004277,2005491,2005489,2006595,2004336,2006633,2002840,2007036,2003198,2004722,2005728,2005796,2003194]

tmp_df_poigraph=df_poigraph[df_poigraph['bm_delivery_area_id'].isin(alist)]
df_area_id=tmp_df_poigraph['bm_delivery_area_id'].drop_duplicates()
```

```python
area_id_list=df_area_id.tolist()
print area_id_list
```

```python
count=0

id_color={'0':'blue','1':'green','2':'purple','3':'yellow','4':'pink','5':'brown','6':'white','7':'maroon','-1':'gray'}

for id in area_id_list:
    print count
    count=count+1
#     if count==3:
#         break
    
    tmp_df_poi=df_poi[df_poi['bm_delivery_area_id']==id]
    tmp_df_area_id=df_poigraph[df_poigraph['bm_delivery_area_id']==id]
    
    import folium
    
    #地图中心点
    map_center_lat=31.102557;
    map_center_lng=121.609862;
    for calss_index in tmp_df_area_id.index :
        tmp_class_id=tmp_df_area_id.at[calss_index,"poi_class_id"]
        map_center_lat =float(tmp_df_area_id.at[calss_index,"poi_class_center_lat"])/1000000
        map_center_lng =float(tmp_df_area_id.at[calss_index,"poi_class_center_lng"])/1000000
    
#     tmp_titles="http://webrd02.is.autonavi.com/appmaptile?lang=zh_cn&size=1&scale=1&style=8&x={x}&y={y}&z={z}"
    map_1 = folium.Map(location=[map_center_lat, map_center_lng],
                       zoom_start=12,
                       tiles='http://webrd02.is.autonavi.com/appmaptile?lang=zh_cn&size=1&scale=1&style=8&x={x}&y={y}&z={z}',
                       attr='Mapbox attribution'
                      )
    
    #商圈可视化
    for calss_index in tmp_df_area_id.index :
        tmp_class_id=tmp_df_area_id.at[calss_index,"poi_class_id"]
        tmp_poi_class_center_lat =float(tmp_df_area_id.at[calss_index,"poi_class_center_lat"])/1000000
        tmp_poi_class_center_lng =float(tmp_df_area_id.at[calss_index,"poi_class_center_lng"])/1000000
        folium.Marker([tmp_poi_class_center_lat, tmp_poi_class_center_lng], popup=str(tmp_class_id),icon=folium.Icon(color='red',icon='info-sign')).add_to(map_1)
        
    
    #POI可视化
    for poi_index in tmp_df_poi.index :
        tmp_poi_class_id=tmp_df_poi.at[poi_index,"poi_class_id"]
        tmp_poi_name=tmp_df_poi.at[poi_index,"poi_name"]
        tmp_poi_lat =float(tmp_df_poi.at[poi_index,"sender_lat"])/1000000
        tmp_poi_lng =float(tmp_df_poi.at[poi_index,"sender_lng"])/1000000
        folium.Marker([tmp_poi_lat, tmp_poi_lng], popup='poi name :'+str(tmp_poi_name)  +',class id:'+str(tmp_poi_class_id),icon=folium.Icon(color=id_color.get(str(int(tmp_poi_class_id))),icon='info-sign')).add_to(map_1)
        
    path=str(id)+'_map.html'
    print path
    map_1.save(path)

```
